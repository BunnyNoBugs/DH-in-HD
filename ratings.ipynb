```python
def calculate_indegree_rating(x, metrics):
    sum = 0
    if x != 'None':
        x = x.split()
        for i in x:
            indegree = metrics[metrics['Id'] == i].iloc[0]['indegree']
            sum += indegree
            sum -= 1
    return sum        
```

```python
research_directors['indegree_rating'] = research_directors.apply(lambda x: calculate_indegree_rating(x['actors'], metrics), axis=1)
```
